# **Teoría Asintótica**

## **Media de una normal**

### **Ejemplo 1**

```r
# Set up
set.seed(123)  # set the seed
N <- 10000
n <- 1:N       # vector: n=1,2,...,N
```

```r
n_dat <- rnorm(n=n, mean=1, sd=0.5)   # Sample from N(1, 0.5^2)
xbar  <- cumsum(n_dat) / n
plot(n, xbar, type = "l")
abline(h=1.01, col="blue", lty=2)
abline(h=0.99, col="blue", lty=2)
```

### **Ejemplo 2**

```r
# Set up
set.seed(123)
N <- 10000          # Número total de lanzadas
n <- 1:N            # vector: n = 1, 2, ..., N; Número de lanzadas
```

```r
# Simulación y gráficos
h <- rbinom(n, size=1, prob=1/2)   # vector: S=0 o 1 con p=1/2
x <- cumsum(h)/n
plot(n, x, type = "l", ylim=c(0,1))
abline(h = 0.52, col = "blue", lty=2)
abline(h = 0.48, col = "blue", lty=2)
```
